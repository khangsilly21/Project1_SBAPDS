### Content-Based Filtering

Chuẩn bị dữ liệu

```python
import pandas as pd 
import numpy as np   
df = pd.read_csv("contentbased_filtering_db.csv")
df.dropna()
df.head(5)
```

```python
df.drop(["Year", "Profitability", 'Worldwide Gross'], axis=1, inplace=True)
df.head()
```

Kết hợp thành 1 cột duy nhất để dễ xử lý

```python
def combine_columns(df, cols):
    features = []  
    
    for i in range(df.shape[0]):
        to_add = ""  
        
        for j in range(len(cols)):
            to_add += str(df[cols[j]][i]) + ' '
        
        features.append(to_add)
    
    return features  
```

```python
df.reset_index(drop=True, inplace=True)
df['combined'] = combine_columns(df, df.columns.tolist())
df.head()
```

Cosine Similatary

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

matrix = CountVectorizer().fit_transform(df["combined"])

cs_matrix = cosine_similarity(matrix)
```

```python
cs_matrix_df = pd.DataFrame(cs_matrix)
cs_matrix_df
```

```python
def getRecommendationFilm(film):
    index = df[df["Film"] == film].index[0]

    scores = list(enumerate(cs_matrix[index]))

    scores = sorted(scores, key=lambda x:x[1], reverse=True)

    scores = scores[1:]

    for movie in scores[:5]:
        mv = df["Film"][movie[0]]
        print(mv)
```

Lấy đề xuất 5 phim dựa theo tên phim đã xem

```python
getRecommendationFilm("Music and Lyrics")
```
