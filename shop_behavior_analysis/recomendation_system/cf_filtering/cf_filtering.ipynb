```python

import pandas as pd 
import numpy as np   
df=pd.read_csv("collaborative_filering_db.csv")
```

```python
users = df.drop(columns="timestamp")
users
```

```python
# Tạo bảng rating với các hàng là userId, các cột là movieId, và giá trị là rating
ratings_matrix = users.pivot_table(index='userId', columns='movieId', values='rating', fill_value=0)

ratings_matrix = ratings_matrix.T
# bo cac rating toan bang 0 (phim ko ai rate)

ratings_matrix
```

```python
# Giả sử ratings_matrix là DataFrame chứa ma trận hiện tại
ratings_matrix = ratings_matrix.replace(0, np.nan)  # Thay thế các giá trị 0 bằng NaN để không tính vào trung bình

# Tính giá trị trung bình mỗi cột, bỏ qua NaN
column_means = ratings_matrix.mean(axis=0, skipna=True)

# Chuẩn hóa ma trận: trừ giá trị trung bình của mỗi cột
normalized_matrix = ratings_matrix.subtract(column_means, axis=1)

# Thay thế NaN (các giá trị chưa có rating) bằng 0
normalized_matrix = normalized_matrix.fillna(0)
# ma tran da chuan hoa
normalized_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Giả sử normalized_matrix là ma trận chuẩn hóa từ bước trước
# Tính toán ma trận tương đồng giữa các user
normalized_matrix=normalized_matrix.T

user_similarity = cosine_similarity(normalized_matrix)

# Chuyển đổi kết quả thành DataFrame để dễ đọc
user_similarity_df = pd.DataFrame(user_similarity, index=normalized_matrix.index, columns=normalized_matrix.index)

# Hiển thị ma trận tương đồng
print("Ma trận tương đồng giữa các user:")
user_similarity_df
```

```python
def get_users_rated_item(movie_id, ratings_matrix):
    """
    Lấy danh sách user đã đánh giá movie_id.
    """
    userss = ratings_matrix[ratings_matrix[movie_id] != 0].index.tolist()
    userss=userss[1:]
    return userss



def predict_rating(user_id, movie_id, ratings_matrix, user_similarity_df, k=2):
    """
    Dự đoán rating cho user_id và movie_id dựa trên KNN.
    
    Args:
    - user_id: ID của user cần dự đoán.
    - movie_id: ID của movie cần dự đoán.
    - ratings_matrix: Ma trận normalized ratings.
    - user_similarity_df: Ma trận tương đồng giữa các user.
    - k: Số lượng user hàng xóm gần nhất.
    
    Returns:
    - predicted_rating: Giá trị dự đoán rating.
    """
    # Lấy danh sách các user đã đánh giá movie_id
    users_rated_item = get_users_rated_item(movie_id, ratings_matrix)

    # Lấy độ tương đồng giữa user_id và các user khác
    similarities = user_similarity_df.loc[user_id, users_rated_item]

    # Sắp xếp độ tương đồng theo thứ tự giảm dần
    top_k_users = similarities.nlargest(k).index.tolist()

    # Lấy normalized ratings của top_k_users cho movie_id
    ratings_top_k_users = ratings_matrix.loc[top_k_users, movie_id]

    # Tính giá trị dự đoán
    numerator = sum(ratings_top_k_users * similarities[top_k_users])
    denominator = sum(abs(similarities[top_k_users]))

    # Tránh chia cho 0
    if denominator == 0:
        return 0

    predicted_rating = numerator / denominator
    return predicted_rating

```

```python
#lay du lieu de lay ten phim tu id
movie_titless= pd.read_csv("movies.csv")
```

```python
def get_top_recommendations(user_id, movie_titles=movie_titless, top_n=10):
    """
    Lấy top N bộ phim mà user có khả năng thích nhất dựa trên rating dự đoán.

    Args:
    - user_id: ID của người dùng.
   
    - movie_titles: DataFrame chứa thông tin về movie (bao gồm movie_id và title).
    - top_n: Số lượng bộ phim muốn lấy.

    Returns:
    - top_movies: Danh sách các bộ phim được đề xuất (bao gồm title và rating dự đoán).
    """
    # predicted_ratings: Dictionary chứa các movie_id và rating dự đoán.
    predicted_ratings = {}
    user_id = user_id
    for movie_id in normalized_matrix.columns:
        if pd.isna(normalized_matrix.loc[user_id, movie_id]) or normalized_matrix.loc[user_id, movie_id] == 0:
            predicted_ratings[movie_id] = predict_rating(user_id, movie_id, normalized_matrix, user_similarity_df, k=20)
    # Bước 1: Chuyển đổi predicted_ratings thành DataFrame để dễ xử lý
    predicted_df = pd.DataFrame(list(predicted_ratings.items()), columns=['movieId', 'predicted_rating'])

    # Bước 2: Sắp xếp danh sách rating theo giá trị rating dự đoán (giảm dần)
    predicted_df = predicted_df.sort_values(by='predicted_rating', ascending=False)

    # Bước 3: Lấy thông tin title tương ứng với các movie_id
    top_movies = predicted_df.merge(movie_titles, how='inner', left_on='movieId', right_on='movieId')

    # Bước 4: Lấy top N kết quả
    top_movies = top_movies.head(top_n)[['title', 'predicted_rating']]

    return top_movies
```
